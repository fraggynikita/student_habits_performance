```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('student_habits_performance.csv')
```

```python
df.describe().T
```

```python
df.isnull().sum()
```

```python
numerical = df.select_dtypes(include='float64')
```

```python
categorical = df.select_dtypes(include='object')
```

```python
rating = df.select_dtypes(include='int64')
```

```python
numerical.head()
```

```python
categorical.drop('student_id', axis = 1, inplace=True)
```

```python
rating.drop('age', axis = 1, inplace=True)
```

```python
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(16, 20))

axes = axes.flatten()

for i, ax in zip(numerical, axes):
    plt.title(f' {i}')
    sns.histplot(df[i], bins=40, kde=True, ax = ax)
    ax.set_title(f' {i}')

plt.tight_layout()
plt.show()
```

```python
for i in categorical:
    print(f''' '{i}' \n-  : {df[i].nunique()} \n  {df[i].unique()}\n''')
```

```python
df.dropna(inplace=True)
```

```python
numerical.corr()
```

```python
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(16, 20))

axes = axes.flatten()

for i, ax in zip(numerical.drop('exam_score', axis=1), axes):
    val = df.groupby(i)['exam_score'].mean()

    sns.scatterplot(x=val.index, y=val.values, ax=ax)
    
    ax.set_ylabel('  ')
    ax.set_title(f'      {i}')

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(16, 20))

for i, ax in zip(categorical, axes.flatten()):
    val = df.groupby(i)['exam_score'].mean()
    bars = val.plot(kind='bar', color=['red', 'pink', 'violet'], ax=ax)
    
    for bar in bars.patches:
        height = bar.get_height()
        ax.annotate(f'{height:.2f}',
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 1),  
                    textcoords="offset points",
                    ha='center', va='bottom')
    
    ax.set_title(f'    {i}')  
    ax.set_xlabel(i)  
    ax.set_ylabel('   ')  

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))
colors = sns.color_palette("husl", len(val))
for i, ax in zip(rating, axes.flatten()):
    val = df.groupby(i)['exam_score'].mean()
    bars = val.plot(kind='bar', color=colors, ax=ax)
    
    for bar in bars.patches:
        height = bar.get_height()
        ax.annotate(f'{height:.2f}',
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 1),  
                    textcoords="offset points",
                    ha='center', va='bottom')
    
    ax.set_title(f'    {i}')  
    ax.set_xlabel(i)  
    ax.set_ylabel('   ')  

plt.tight_layout()
plt.show()
```
